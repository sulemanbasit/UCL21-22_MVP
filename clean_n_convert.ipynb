```python
# Champions League stages mapping
champions_league_stages = {
    "Group Stage": ["Leipzig", "Club Brugge", "Porto", "Milan", "Dortmund", "Beşiktaş", "Sheriff", 
                    "Shakhtar Donetsk", "Barcelona", "Dynamo Kyiv", "Atalanta", "Young Boys", 
                    "Sevilla", "Wolfsburg", "Zenit", "Malmö"],
    "Round of 16": ["Ajax", "Inter", "Sporting CP", "Juventus", "Salzburg", "Man. United", "LOSC", "Paris"],
    "Quarter Finals": ["Benfica", "Bayern", "Atlético", "Chelsea"],
    "Semi Finals": ["Villarreal", "Man. City"],
    "Finals": ["Liverpool"],
    "Winner": ["Real Madrid"]
}
```

```python
# Function to get the Champions League stage of a club
def get_cl_stage(club):
    for stage, clubs in champions_league_stages.items():
        if club in clubs:
            return stage
    return "Not Participated"
```

```python
# Add the column to player_data
clean_filled_player_data["Champions_League_Stage"] = clean_filled_player_data["club"].apply(get_cl_stage)
club_data["Champions_League_Stage"] = club_data['club'].apply(get_cl_stage)
```

```python
# For Min/Goals
clean_filled_player_data["min/goals"] = (clean_filled_player_data['minutes_played'] / clean_filled_player_data['goals'].replace(0, np.nan)).fillna(0)

# For Min/Assists
clean_filled_player_data["min/assists"] = (clean_filled_player_data['minutes_played'] / clean_filled_player_data['goals'].replace(0, np.nan)).fillna(0)

# For Goal Involvement
clean_filled_player_data["goal_involvement"] =  clean_filled_player_data['goals'] + clean_filled_player_data['assists']

# For Shot Accuracy
clean_filled_player_data["shot_accuracy"] = ((clean_filled_player_data['on_target'] + clean_filled_player_data['blocked']) / clean_filled_player_data['total_attempts'].replace(0, np.nan)).fillna(0)

# For Min/(G+A)
clean_filled_player_data["min/(g+a)"] = (clean_filled_player_data['minutes_played'] / (clean_filled_player_data["goal_involvement"]).replace(0, np.nan)).fillna(0)

# For Distance Covered/Match Played
clean_filled_player_data["distance/match"] = (clean_filled_player_data['distance_covered'] / clean_filled_player_data['match_played'].replace(0, np.nan)).fillna(0)

# For Successful Tackle Percent
clean_filled_player_data['tackle_accuracy'] = (clean_filled_player_data['t_won'] / clean_filled_player_data['tackles'].replace(0, np.nan)).fillna(0)

# For Tackle Ratio
clean_filled_player_data['tackle_ratio'] = (clean_filled_player_data['t_won'] / clean_filled_player_data['t_lost'].replace(0, np.nan)).fillna(0)

# Saved:Conceded
clean_filled_player_data["saved/conceded"] = (clean_filled_player_data['saved'] / clean_filled_player_data['conceded'].replace(0, np.nan)).fillna(0)

# Cleansheets/Match Played
clean_filled_player_data["cleansheets/match"] = (clean_filled_player_data['cleansheets'] / clean_filled_player_data['match_played'].replace(0, np.nan)).fillna(0)
```

```python
# =========================
# Function to Calculate Player Points
# =========================

def calculate_player_points(row):
    position = row['position']
    if position in position_weights:
        total_points = sum(
            row.get(metric, 0) * weight for metric, weight in position_weights[position].items()
        )
        total_points += cl_stage_points.get(row['Champions_League_Stage'], 0)
        return total_points
    return 0

# =========================
# Function to Calculate Club Points
# =========================
def calculate_club_points(row):
    total_points = sum(
        row.get(metric, 0) * weight for metric, weight in club_weights.items()
    )
    total_points += cl_stage_points.get(row['Champions_League_Stage'], 0)
    return total_points
```

```python
clean_filled_player_data['Total Points'] = clean_filled_player_data.apply(calculate_player_points, axis=1)
club_data['Total Points'] = club_data.apply(calculate_club_points, axis=1)
```
